```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt




codes = pd.read_csv("Wifi-data.tsv", sep="\t")
unique= codes['zip'].value_counts().rename_axis('Zip Codes').to_frame(name='Wifi Count')

#data = unique.value_counts().rename_axis('Zip Codes').to_frame(name='Store Count')
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)


#print(data)

unique.plot.bar(figsize=(20, 15))

unique.head(10)
```

When it comes to public wifi the top neighborhoods to find it free and easy to access are Braddock, Mt. Oliver, and the Central Business District. Coming in at the top is Braddock with 18 public locations followed by Mt Oliver with a total of 13. Finally in third we have the Central Business District with a total of 9 different spots.
